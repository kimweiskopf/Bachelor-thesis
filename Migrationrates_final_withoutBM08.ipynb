# Migrationsraten

```python
#functions to calculate the rates and loctaions of the simulations
T=np.pi*2
def migrationrate_planets_r(path,start,stop):
    t,a =np.loadtxt(str(path)+'/orbit0.dat', usecols = (0,2), unpack=True )
    t1,a1 =np.loadtxt(str(path)+'/orbit1.dat', usecols = (0,2), unpack=True)
    
    delta_a= a[stop]-a[start]
    delta_t=t[stop]-t[start]

    return  (a[stop]-a[start])*0.5+a[start]

def migrationrate_planets_m(path,start,stop):
    t,a =np.loadtxt(str(path)+'/orbit0.dat', usecols = (0,2), unpack=True )
    t1,a1 =np.loadtxt(str(path)+'/orbit1.dat', usecols = (0,2), unpack=True)
    
    delta_a= a[stop]-a[start]
    delta_t=t[stop]-t[start]
    
    return delta_a/delta_t
```
